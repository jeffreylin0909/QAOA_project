```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
corrcoef_list = []

for instance_id in range(20):
    grad_var = pd.read_csv(f"grad_var/{instance_id}.csv", header=None)
    grad_var = np.array(grad_var[0])
    weighted_degree = pd.read_csv(f"weighted_degree/{instance_id}.csv", header=None)
    weighted_degree = np.array(weighted_degree[0])
    corrcoef_list.append(np.corrcoef(grad_var, weighted_degree)[0][1])
    plt.scatter(grad_var, weighted_degree)  # 產生散佈圖
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("weighted degree relation with beta gradient variance of nodes")
    plt.show()
    print(f"correlation coefficient: {corrcoef_list[-1]}")
    
```

```python
np.mean(corrcoef_list)
```
